```python
# create data step
import numpy as np
import matplotlib.pyplot as plt

X = np.array([
    [2.5, 2.4],
    [0.5, 0.7],
    [2.2, 2.9],
    [1.9, 2.2],
    [3.1, 3.0],
    [2.3, 2.7],
    [2.0, 1.6],
    [1.0, 1.1],
    [1.5, 1.6],
    [1.1, 0.9]
])
```

```python
mean = np.mean(X, axis=0)
X_centered = X - mean

#center data by mean
```

```python
cov_matrix = np.cov(X_centered.T)

#compute cov matrix, tells us spread and variance of data
```

```python
#calculate eigen values and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
```

```python
sorted_idx = np.argsort(eigenvalues)[::-1]
top_eigenvector = eigenvectors[:, sorted_idx[0]]

X_reduced = X_centered @ top_eigenvector
```

```python
plt.scatter(X_centered[:, 0], X_centered[:, 1], label='Original Data')
for point, proj in zip(X_centered, X_reduced):
    proj_point = proj * top_eigenvector
    plt.plot([point[0], proj_point[0]], [point[1], proj_point[1]], 'r--')

plt.axline((0, 0), slope=top_eigenvector[1] / top_eigenvector[0], color='g', label='PC1')
plt.title('PCA Projection onto First Principal Component')
plt.axis('equal')
plt.legend()
plt.show()
```
