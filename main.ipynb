```python
from Linear_Algorithms import SimpleLinearRegression
import numpy as np
import matplotlib.pyplot as plt

num_points = 100

noise_std = 20

X = np.random.rand(num_points) * 100

y = 3.12345 * X + 20 + np.random.normal(0, noise_std, num_points)
```

```python
plt.scatter(X, y, color='blue', label='Data points')

plt.xlabel('X')
plt.ylabel('y')

plt.legend()
plt.show()
```

```python
m, b = SimpleLinearRegression(X, y)

print(f"Slope: {m:.5} \ny-intercept: {b:.2}")
```

```python
X_line = np.linspace(X.min(), X.max(), 100)
y_line = m * X_line + b


plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_line, y_line, color='red', label='Simple Linear Regression')
for i in range(len(X)):
    plt.plot([X[i], X[i]], [y[i], m * X[i] + b], color='gray', linestyle='--')
plt.xlabel('X')
plt.ylabel('y')

plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

m1 = 2
m2 = 3
b = 5

noise_std = 1

num_points = 100

X1 = np.random.rand(num_points) * 100
X2 = np.random.rand(num_points) * 50

noise = np.random.normal(0, noise_std, num_points)

y = m1 * X1 + m2 * X2 + b + noise

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X1, X2, y, color='blue', label='Data Points')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')
plt.legend()
plt.show()
```

```python
from Linear_Algorithms import MutipleLinearRegression
m, b = MutipleLinearRegression([X1, X2], y)
print(f"Slope: {m}\ny-intercept: {b:.4f}")
```

```python
# Create a grid of X1 and X2 values
X1_grid, X2_grid = np.meshgrid(np.linspace(X1.min(), X1.max(), 10),
                               np.linspace(X2.min(), X2.max(), 10))

# Compute the corresponding y values for the plane
y_plane = m[0] * X1_grid + m[1] * X2_grid + b


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1_grid, X2_grid, y_plane, color='red', alpha=0.3, rstride=100, cstride=100)

ax.scatter(X1, X2, y, color='blue', label='Data Points')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')
plt.legend()
plt.show()
```
